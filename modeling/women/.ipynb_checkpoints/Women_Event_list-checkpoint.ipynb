```python

data = {
    'Event': ['Balance Beam', 'Balance Beam', 'Vault', 'Uneven Bars', 'Uneven Bars', 'IAA', 'Floor', 'Floor'],
    'Gymnast': ['Simone Biles', 'Joscelyn Roberson', 'Simone Biles', 'Shilese Jones', 'Zoe Miller', 'Jordan Chiles', 'Simone Biles', 'Ashlee Sullivan']
}

gymnastics_df = pd.DataFrame(data)

gymnastics_df
```

```python
gymnastics_df['Gymnast'].value_counts()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df=gymnastics_df.copy()

# Plotting a bar chart with counts for each athlete
plt.figure(figsize=(10, 6))
sns.countplot(x='Gymnast', data=df, palette='viridis')
plt.title('Athlete Medal Counts for Events')
plt.xlabel('Gymnast')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Plotting a grouped bar chart
plt.figure(figsize=(12, 6))
sns.countplot(x='Gymnast', hue='Event', data=df, palette='Set2')
plt.title('Gymnast Participation in Events')
plt.xlabel('Gymnast')
plt.ylabel('Count')
plt.legend(title='Event', bbox_to_anchor=(1, 1))

# Display the plot
plt.show()
```

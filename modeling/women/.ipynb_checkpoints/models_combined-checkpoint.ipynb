```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(22, 12))

# Plot accuracy
plt.subplot(1, 2, 1)
sns.barplot(x='Classifier', y='Accuracy', hue='Event', data=df)
plt.title('Accuracy for Each Classifier')
plt.ylim(0.9, 1.001) 

# Move legends outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Plot F2 score
plt.subplot(1, 2, 2)
sns.barplot(x='Classifier', y='F2-Score', hue='Event', data=df)
plt.title('F2-Score for Each Classifier')
plt.ylim(0.3, 1.01)  

# Move legends outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.tight_layout()
plt.show()
```

```python
fxusa=fx[fx['nation']=='USA']
vtusa=vt[vt['Nation']=='USA']
bbusa=bb[bb['nation']=='USA']
ubusa=ub[ub['nation']=='USA']
```

```python
fxusa['apparatus']='fx'
vtusa['apparatus']='vt'
bbusa['apparatus']='bb'
ubusa['apparatus']='ub'
```

```python
fxusa=fxusa.drop(columns=['Rank', 'nation', 'round', 'D Score', 'E Score', 'Pen.'])
ubusa=ubusa.drop(columns=['Rank', 'nation', 'round', 'D Score', 'E Score', 'Pen.'])
bbusa=bbusa.drop(columns=['Rank', 'nation', 'round', 'D Score', 'E Score', 'Pen.'])
```

```python
vtusa=vtusa.drop(columns=['Rank', 'Nation', 'D', 'E', 'ND', 'Total', 'D2', 'E2', 
                          'ND2', 'Total2', 'round'])
```

```python
vt=vtusa.copy()
fx=fxusa.copy()
ub=ubusa.copy()
bb=bbusa.copy()
```

```python
vt=vt.drop_duplicates()
fx=fx.drop_duplicates()
ub=ub.drop_duplicates()
bb=bb.drop_duplicates()
```

```python
vt['Athlete']=vt['Athlete'].str.lower()
fx['Name']=fx['Name'].str.lower()
bb['Name']=bb['Name'].str.lower()
ub['Name']=ub['Name'].str.lower()
```

```python
fx20 = fx[fx['year'] < 2021]
vt20 = vt[vt['year'] < 2021]
ub20 = ub[ub['year'] < 2021]
bb20 = bb[bb['year'] < 2021]

fx20 = fx20[fx20['year'] > 2016]
vt20 = vt20[vt20['year'] > 2016]
ub20 = ub20[ub20['year'] > 2016]
bb20 = bb20[bb20['year'] > 2016]
```

```python
fx20
```

```python
vt20
```

```python
column_name_mapping = {'Average': 'Total', 'Athlete': 'Name'}

# Rename columns using the dictionary
vt20.rename(columns=column_name_mapping, inplace=True)
```

```python
fx20['Name'].value_counts()
indices_to_keep = fx20.groupby('Name')['Total'].idxmax()

# Filter the DataFrame based on the identified indices
df_largest_scores = fx20.loc[indices_to_keep]

# Reset the index if needed
df_largest_scores.reset_index(drop=True, inplace=True)
```

```python
fx=df_largest_scores.copy()
```

```python
bb20
```

```python
df = pd.concat([fx, ub20, bb20, vt20], ignore_index=True)
```

```python
df
```

```python
df_aggregated = df.groupby(['Name', 'apparatus'])['Total'].sum().reset_index()

# Pivot the aggregated DataFrame
pivoted_df = df_aggregated.pivot(index='Name', columns='apparatus', values='Total')

# Reset the index if needed
pivoted_df.reset_index(inplace=True)
```

```python
pivoted_df.fillna(0, inplace=True)
```

```python
df=pivoted_df.copy()
```

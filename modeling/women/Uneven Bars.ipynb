# Base line using ZeroR

```python

X = df.drop(columns=['medal'])  
y = df['medal']  

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=90)

# Initialize the ZeroR classifier
zero_r_clf = DummyClassifier(strategy='most_frequent', random_state=42)

zero_r_clf.fit(X_train, y_train)

# Make predictions test data
y_pred = zero_r_clf.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("ZeroR Classifier")
print(f"Accuracy: {accuracy:.4f}")
print(f"Classification Report:\n{report}")
print(f"Confusion Matrix:\n{conf_matrix}")
```

# Model Comparison using Grid Search and k-folds cross validation

```python

# Assuming df is your DataFrame
y = df['medal']
X = df.drop(columns=['medal'])

# Initialize classifiers
classifiers = {
    "Random Forest": RandomForestClassifier(),
    "AdaBoost": AdaBoostClassifier(),
    "SVM": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Naive Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Bagging": BaggingClassifier(),
    "Extra Trees": ExtraTreesClassifier(),
}

results = {}

# Initialize StratifiedKFold for k-fold cross-validation
k_folds = 5
stratified_kfold = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

# Iterate through each classifier
for name, clf in classifiers.items():
    # Create a pipeline with StandardScaler for classifiers that require it
    if name in ["SVM", "K-Nearest Neighbors", "Neural Network", "Logistic Regression"]:
        clf = make_pipeline(StandardScaler(), clf)

    # Define the parameter grid for the current classifier
    param_grid = {}
    if name == "Random Forest":
        param_grid = {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    elif name == "AdaBoost":
        param_grid = {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2]
        }
    # Add similar blocks for other classifiers

    # Perform grid search with 5-fold cross-validation
    grid_search = GridSearchCV(clf, param_grid, cv=stratified_kfold, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X, y)

    # Get the best estimator from grid search
    best_clf = grid_search.best_estimator_

    # Initialize variables to store results
    accuracies = []

    # Iterate over folds for evaluation
    for fold, (train_index, test_index) in enumerate(stratified_kfold.split(X, y)):
        # Split the data into training and testing sets for this fold
        X_train_fold, X_val_fold = X.iloc[train_index], X.iloc[test_index]
        y_train_fold, y_val_fold = y.iloc[train_index], y.iloc[test_index]

        # Train the model for this fold
        best_clf.fit(X_train_fold, y_train_fold)

        # Make predictions on the validation data for this fold
        y_pred = best_clf.predict(X_val_fold)

        # Evaluate the model's performance for this fold
        accuracy = accuracy_score(y_val_fold, y_pred)
        accuracies.append(accuracy)

        print(f"\nClassifier: {name}, Fold {fold + 1}")
        print(f"Accuracy: {accuracy:.4f}")

    # Calculate and print the average accuracy across folds
    average_accuracy = sum(accuracies) / k_folds
    print(f"\nClassifier: {name}, Average Accuracy across Folds: {average_accuracy:.4f}")

    # Make predictions on the test data
    y_pred_test = best_clf.predict(X_test)

    # Evaluate the model's performance on the test set
    accuracy_test = accuracy_score(y_test, y_pred_test)
    class_report_test = classification_report(y_test, y_pred_test)
    conf_matrix_test = confusion_matrix(y_test, y_pred_test)

    # Store the results for later use
    results[name] = {
        "best_clf": best_clf,
        "accuracy_test": accuracy_test,
        "class_report_test": class_report_test,
        "conf_matrix_test": conf_matrix_test,
    }
    
    # Find the best model based on the maximum test set accuracy
    best_model_name = max(results, key=lambda k: results[k]['accuracy_test'])
    best_model_accuracy = results[best_model_name]['accuracy_test']

    # Iterate over the remaining models to check if there's a better one
    for name, result in results.items():
        if result['accuracy_test'] > best_model_accuracy:
            best_model_name = name
            best_model_accuracy = result['accuracy_test']

    # Print the overall best model and its test set accuracy
    print(f"\nOverall Best Model: {best_model_name}")
    print(f"Test Set Accuracy: {best_model_accuracy:.4f}")


# Print the results
for name, result in results.items():
    print(f"\nClassifier: {name}")
    print(f"Best Estimator: {result['best_clf']}")
    print(f"Test Set Accuracy: {result['accuracy_test']:.4f}")
    print(f"Classification Report:\n{result['class_report_test']}")
    print(f"Confusion Matrix:\n{result['conf_matrix_test']}")
```

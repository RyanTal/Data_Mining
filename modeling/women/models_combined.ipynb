```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 5))

# Plot F2-Score for each event and classifier
ax = sns.barplot(x='Classifier', y='Accuracy', hue='Event', data=df)
plt.title('Accuracy for Each Event and Classifier')
plt.xlabel('Classifier')
plt.ylabel('Accuracy')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set y-axis limits to start at 0.7
ax.set_ylim(0.90, 1.01)

# Move the legend outside of the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
# Pivot the DataFrame for the heatmap
heatmap_data = df.pivot(index='Classifier', columns='Event', values='F2-Score')

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".2f", linewidths=.5, cbar_kws={'label': 'F2-Score'})

# Set labels and title
plt.xlabel('Event')
plt.ylabel('Classifier')
plt.title('F2-Score Heatmap')

# Show the plot
plt.show()
```

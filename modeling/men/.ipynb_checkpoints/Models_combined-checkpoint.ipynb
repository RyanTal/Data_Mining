```python
import pandas as pd

data = {
    'Event': ['Parallel Bars']*7,
    'Classifier': ['Random Forest', 'AdaBoost', 'SVM', 'K-Nearest Neighbors', 'Decision Tree', 'Naive Bayes', 'Neural Network'],
    'Accuracy': [0.9752, 0.9876, 0.9689, 0.9689, 0.9938, 0.7702, 0.9689],
    'Precision': [0.97, 0.99, 0.97, 0.97, 1, 1, 0.97],
    'Recall': [1, 1, 1, 1, 0.99, 0.76, 1],
    'F1-Score': [0.98, 0.99, 0.98, 0.98, 1, 0.87, 0.98],
    'Support': [161, 161, 161, 161, 161, 161, 161],
    'F2-Score': [0.2381, 0.6522, 0, 0, 0.9615, 0.4032, 0]
}

pb = pd.DataFrame(data)
pb
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 5))

# Plot F2-Score for each event and classifier
sns.barplot(x='Classifier', y='F2-Score', hue='Event', data=df)
plt.title('F2-Score for Each Event and Classifier')
plt.xlabel('Classifier')
plt.ylabel('F2-Score')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Move the legend outside of the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 5))

# Plot F2-Score for each event and classifier
ax = sns.barplot(x='Classifier', y='Accuracy', hue='Event', data=df)
plt.title('Accuracy for Each Event and Classifier')
plt.xlabel('Classifier')
plt.ylabel('Accuracy')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set y-axis limits to start at 0.7
ax.set_ylim(0.90, 1.01)

# Move the legend outside of the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```

```python
# Pivot the DataFrame for the heatmap
heatmap_data = df.pivot(index='Classifier', columns='Event', values='F2-Score')

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a heatmap
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt=".2f", linewidths=.5, cbar_kws={'label': 'F2-Score'})

# Set labels and title
plt.xlabel('Event')
plt.ylabel('Classifier')
plt.title('F2-Score Heatmap')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(x='F2-Score', y='Accuracy', hue='Classifier', data=df, s=100)

# Set x-axis and y-axis limits
plt.xlim(0.4, 1.01)  # Adjust the starting value for the x-axis
plt.ylim(0.7, 1.01)  # Adjust the starting value for the y-axis

# Set labels and title
plt.xlabel('F2-Score')
plt.ylabel('Accuracy')
plt.title('Scatter Plot of Accuracy vs F2-Score')

# Add legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select numeric columns for the pair plot
numeric_columns = [ 'Accuracy', 'Precision', 'Recall', 'F1-Score', 'F2-Score']

# Add the 'Event' column to distinguish different events
pairplot_data = pd.concat([df['Classifier'], df[numeric_columns]], axis=1)

# Create a pair plot
g = sns.pairplot(pairplot_data, hue='Classifier', markers='o', palette='viridis')

# Set the y-axis limit for all subplots
g.set(ylim=(0.7, 1.01))

plt.show()
```

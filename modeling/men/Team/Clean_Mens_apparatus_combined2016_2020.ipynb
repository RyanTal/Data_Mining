```python
iaa = pd.read_csv('../../../Combine_Data/men/iaa2020.csv')
iaa= iaa[iaa['Nation']=='USA']
```

```python
iaa=iaa.drop(columns=['Rank', 'Nation', 'AA', 'round', 'year'])
```

```python
iaa['Athlete'].value_counts()
```

```python
iaa['Athlete'] = iaa['Athlete'].replace('sam mikulak', 'samuel mikulak')
```

```python
import pandas as pd



# Convert scores columns to numeric to handle any non-numeric values
score_columns = ['FX', 'PH', 'SR', 'VT', 'PB', 'HB']
iaa[score_columns] = iaa[score_columns].apply(pd.to_numeric, errors='coerce')

# Group by 'Athlete' and aggregate using the maximum score for each column
aggregated_df = iaa.groupby('Athlete').agg({
    'FX': 'max',
    'PH': 'max',
    'SR': 'max',
    'VT': 'max',
    'PB': 'max',
    'HB': 'max'
}).reset_index()


```

```python
iaa=aggregated_df.copy()
```

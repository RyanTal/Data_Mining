```python
import tabula
import pandas as pd
from tabula import read_pdf

start_page = 283
end_page = 288


# Path to your PDF file
pdf_path = "resultbook.pdf"

pages_to_scrape = [str(page) for page in range(start_page, end_page + 1)]

# Use tabula to read tables from the specified range of pages
dfs = read_pdf(pdf_path, pages=pages_to_scrape, multiple_tables=True)

# Concatenate the DataFrames if there are multiple tables on different pages
df = pd.concat(dfs)


# Display the resulting DataFrame
# print(df)
df
```

```python
df=df.iloc[2:]
```

```python

df.rename(columns={'Unnamed: 1': 'Rank', 'Unnamed: 4': 'Pen.', 'Unnamed: 5': 'Total'}, inplace=True)
```

```python
df=df.iloc[0:221]
```

```python
df=df.drop(columns=['Unnamed: 0'])
```

```python
df=df.drop(columns=['Unnamed: 2'])
```

```python
df['Pen.']=df['Pen.'].fillna(0)
```

```python
fx=df.copy()
```

```python
fx.info()
```

```python
fx=fx.dropna()
```

```python
fx.rename(columns={'Unnamed: 3': 'Name'}, inplace=True)
```

```python
fx[['Last Name', 'First Name']] =  fx['Name'].str.rsplit(' ', 1, expand=True)
```

```python
fx=fx.drop(columns=['Name'])
```

```python
fx['Name']=fx['First Name'] + ' ' + fx['Last Name']
```

```python
fx=fx.drop(columns=['First Name', 'Last Name'])
```

```python
fx
```

```python
# fx.to_csv('wrld15champ_fx.csv', index=False)
```

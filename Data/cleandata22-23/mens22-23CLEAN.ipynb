```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_2022_2023.csv')
```

```python
df["Competition"].value_counts()
```

```python
df.info()
```

```python
df["Penalty"] = df["Penalty"].fillna(0)
```

```python
df.info()
```

```python
df["Rank"] = df["Rank"].fillna(0)
```

```python
df["Country"] = df["Country"].fillna("None")
```

```python
df["LastName"] = df["LastName"].fillna("None")
df["FirstName"] = df["FirstName"].fillna("None")
df["D_Score"] = df["D_Score"].fillna(0)
df["E_Score"] = df["E_Score"].fillna(0)
```

```python
df.info()
```

```python
df["Total_Score"] = df["D_Score"] + df["E_Score"] - df["Penalty"]
```

```python
df.info()
```

```python
df["Score"] = df["Score"].fillna(0)
```

```python
df.isnull().sum()
```

```python
df["Name"] = df["FirstName"] + " " + df["LastName"] 
```

```python
df = df.drop(columns=["LastName", "FirstName"])
```

```python
df["Date"]
```

```python
df[['Start Date', 'End Date']] = df['Date'].str.split('-', expand=True)
```

```python
df = df.drop(columns=["Start Date"])
```

```python
# Convert the 'End Date' column to datetime format with automatic format inference
df['End Date'] = pd.to_datetime(df['End Date'], errors='coerce', infer_datetime_format=True)

```

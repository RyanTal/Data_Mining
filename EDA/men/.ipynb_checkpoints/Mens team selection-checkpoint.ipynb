# Hyper-Heuristics Averages:

```python
numeric_columns = hhavg.select_dtypes(include=['number']).columns

# Get the top 3 values for each column (excluding 'Athlete')
top3_values = hhavg[numeric_columns].apply(lambda x: x.nlargest(3))

# Calculate the sum of the top three for each column
sum_of_top3 = top3_values.sum()

# Calculate the sum of the top 3 values across all columns
total_sum_of_top3 = sum_of_top3.sum()

# Display the top 3 values for each column
print("Top 3 values for each column:")
print(top3_values)

# Display the sum of the top 3 values for each column
print("\nSum of top 3 values for each column:")
print(sum_of_top3)

# Display the total sum of the top 3 values across all columns
print("\nUsing average score of athletes in HH:", total_sum_of_top3)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fxm = pd.read_csv('../../Combine_Data/men/fx_encoded.csv')
```

```python
df=fxm.copy()
```

```python
sns.countplot(x='medal', data=df)
plt.title('medal')
plt.show()
```

```python
df.describe()
```

```python
df.hist(figsize=(12, 10))
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
df.info()
```

```python
selected_features = ['D', 'E', 'ND', 'Total', 'medal']
sns.pairplot(df[selected_features], hue='medal')
plt.show()
```

```python
numeric_features = ['D', 'E', 'ND', 'Total']
for feature in numeric_features:
    sns.boxplot(x='medal', y=feature, data=df)
    plt.title(f'Boxplot of {feature} by Medal')
    plt.show()
```

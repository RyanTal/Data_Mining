```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Assuming df is your DataFrame

# Remove scores less than 5 for each apparatus
for apparatus in df.columns[1:-1]:
    df = df[df[apparatus] >= 5]

# Melt the DataFrame to long format for easier plotting
df_long = pd.melt(df, id_vars=['Athlete'], var_name='Apparatus', value_name='Score')

# Set the style of seaborn
sns.set(style="whitegrid")

# Manually assign colors to the highlighted athletes
highlighted_colors = {'asher hong': 'red', 'yul moldauer': 'blue', 
                      'fred richard': 'green', 'khoi young': 'purple', 
                      'vitaliy guimaraes': 'orange'}

# Create a dropdown widget for selecting the apparatus
apparatus_dropdown = widgets.Dropdown(
    options=df.columns[1:-1],
    value=df.columns[1],
    description='Apparatus:'
)

# Update the plot_scores function to take widget values as arguments
@interact(apparatus=apparatus_dropdown)
def plot_scores(apparatus):
    plt.figure(figsize=(8, 6))

    # Assign colors based on athlete
    df_long['Color'] = df_long['Athlete'].map(lambda x: highlighted_colors.get(x, 'grey'))

    # Create a scatter plot for the selected apparatus
    sns.scatterplot(x='Athlete', y='Score', hue='Color', data=df_long[df_long['Apparatus'] == apparatus], s=100)

    plt.title(f'Scores on {apparatus} for Each Athlete')
    plt.xlabel('Athlete')
    plt.ylabel('Score')

    # Manually create the legend
    legend_labels = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10, label=athlete) for athlete, color in highlighted_colors.items()]
    plt.legend(handles=legend_labels, bbox_to_anchor=(1.05, 1), loc='upper left', title='Athlete')
    
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

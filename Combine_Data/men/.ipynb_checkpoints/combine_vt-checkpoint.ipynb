```python
import pandas as pd
import numpy as np
```

```python
wc19melb = pd.read_csv('../../Cleaning/qual_events/men/wc19bmelb_vt.csv')
world18 = pd.read_csv('../../Cleaning/qual_events/men/world18championship_vt.csv')

world19 = pd.read_csv('../../Cleaning/qual_events/men/world_championship2019vt.csv')
uni19 = pd.read_csv('../../Cleaning/qual_events/men/uni2019vt.csv')
winter20cup = pd.read_csv('../../Cleaning/qual_events/men/winter20cup_vt.csv')
wc20baku = pd.read_csv('../../Cleaning/qual_events/men/wc20baku_vt.csv')
eu19vt = pd.read_csv('../../Cleaning/qual_events/men/european_championship2019vt.csv')
world19qual = pd.read_csv('../../Cleaning/qual_events/men/world_championship2019vt_qual.csv')

```

```python
new_column_order = ['Rank', 'Athlete', 'Nation', 'D', 'E', 'ND', 'Total', 'D.1', 'E.1',
       'ND.1', 'Total.1', 'Average', 'year', 'round']

# Create a new DataFrame with the desired column order
world19qual = world19qual[new_column_order]
```

```python
new_column_names = {'avg': 'Average'}
uni19 = uni19.rename(columns=new_column_names)
```

```python
new_column_names = {'Total': 'Average'}
winter20cup = winter20cup.rename(columns=new_column_names)
```

```python
new_column_names = {'Score': 'Total', 'Score.1': 'Total.1'}
winter20cup = winter20cup.rename(columns=new_column_names)
```

```python
vt = pd.concat([wc19melb, world18, world19, uni19, winter20cup, wc20baku, eu19vt,
                world19qual])
```

```python
vt = vt.drop(columns=['round'])
```

```python
vt['Nation'].fillna('none', inplace=True)
```

```python
vt.info()
```

```python
vt
```

```python
values_to_replace = ['DNS', 'DNF']
replacement_value = 0

vt.replace(values_to_replace, replacement_value, inplace=True)
```

```python
vt
```

```python
vt=vt.drop(columns=['Average'])
```

```python
vt.info()
```

```python
columns_to_convert = ['D.1', 'E.1', 'Total.1']

# Loop through each column and perform the conversion
for column in columns_to_convert:
    vt[column] = pd.to_numeric(vt[column], errors='coerce')
```

```python

vt['Average'] = vt[['Total', 'Total.1']].mean(axis=1)
```

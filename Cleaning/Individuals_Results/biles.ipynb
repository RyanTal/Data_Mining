```python
# Define a function to combine 'date' and 'Year' columns into datetime objects
def combine_date_year(row):
    date_str = row['Date'] + f" {row['Year']}"
    return datetime.strptime(date_str, "%b %d %Y")

# Apply the function to create a new 'date' column with datetime objects
df['Date'] = df.apply(combine_date_year, axis=1)
```

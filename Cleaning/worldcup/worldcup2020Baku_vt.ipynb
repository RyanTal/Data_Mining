```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../scraping/women/worldcup20Baku/worldcup2020Baku_vt.csv')
```

```python
df
```

```python
odds = df[1::2].reset_index(drop=True)
```

```python
odds
```

```python
# Create a new DataFrame by selecting only the odd rows
evens = df[::2].reset_index(drop=True)

```

```python
odds.head()
```

```python
odds = odds.drop(columns=["Rank", "Athlete", "Nation"])
```

```python
column_mapping = {
    'D': 'D2',
    'E': 'E2',
    'ND': 'ND2',
    'Total': 'Total2'
    
}

# Use the rename method to rename the columns
odds.rename(columns=column_mapping, inplace=True)
```

```python
odds['ND2'].fillna(0, inplace=True)
```

```python
evens.head()
```

```python
evens = evens.drop(columns=["Average"])
```

```python
evens.head()
```

```python
evens['ND'].fillna(0, inplace=True)
```

```python
result_df = pd.concat([evens, odds], axis=1)
```

```python
evens
```

```python
result_df
```

```python
df = result_df.copy()
```

```python
df = df.replace(['DNF', 'DNS'], 0)
```

```python
df
```

```python
df.info()
```

```python
df['Nation'].unique()
```

```python
nation_to_noc = {
    'Slovenia': 'SLO',
    'France': 'FRA',
    'Azerbaijan': 'AZE',
    'Canada': 'CAN',
    'Mexico': 'MEX',
    'Ukraine': 'UKR',
    'Croatia': 'CRO',
    'Turkey': 'TUR',
    'Russia': 'RUS',
    'New Zealand': 'NZL',
    'Poland': 'POL',
    'Finland': 'FIN',
    'Belarus': 'BLR',
    'Great Britain': 'GBR',
    'Georgia': 'GEO'
}

# Replace the 'Nation' column values with their corresponding NOC codes
df['Nation'] = df['Nation'].map(nation_to_noc)
```

```python
df
```

```python
# Columns to convert to float
float_columns = ['D', 'E', 'ND', 'Total', 'D2', 'E2', 'ND2', 'Total2', 'Average']

# Loop through each column and convert to float
for column in float_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')


```

```python
df
```

```python
df.info()
```

```python
df['Rank'] = pd.to_numeric(df['Rank'], errors='coerce', downcast='integer')
```

```python
df.info()
```

```python
df.head()
```

```python
column_mapping = {
    'Total': 'Score',
    'Total2': 'Score2',
    'Average': 'Total'

    
}

# Use the rename method to rename the columns
df.rename(columns=column_mapping, inplace=True)
```

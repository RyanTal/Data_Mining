```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../scraping/women/qual2008/scrape2008qual_team.csv')
```

```python
df.tail(50)
```

```python
# Select rows 0 and 5 and create a new DataFrame
new_df = pd.concat([df.iloc[0], df.iloc[7], df.iloc[14], df.iloc[28], df.iloc[35],
                   df.iloc[42], df.iloc[49], df.iloc[56], df.iloc[63],
                   df.iloc[70], df.iloc[77]], axis=1).transpose()

# Drop rows 0 and 5 from the original DataFrame
df = df.drop([0, 7, 14, 28, 35, 42, 49, 56, 63, 70, 77])

# Reset the index of the original DataFrame
df = df.reset_index(drop=True)
```

```python
# Extract data within parentheses and store it in a new column
df["Country"] = df['Team'].str.extract(r'\((.*?)\)')

df['Team'] = df['Team'].apply(lambda x: x.split('(')[0].strip())
df.info()
```

```python
df
```

```python
df.fillna(0, inplace=True)
df
```

```python
df["Name"] = df["Team"]
df=df.drop(columns=["Team"])
df
```

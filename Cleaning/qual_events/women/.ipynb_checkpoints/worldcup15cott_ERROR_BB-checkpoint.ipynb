```python
import pandas as pd
import numpy as np
```

```python
bb = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_bb.csv')
bbq = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_qual_bb.csv')

ub = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_ub.csv')
ubq = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_qual_ub.csv')

vt = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_vt.csv')
# vtq = pd.read_csv('../../scraping/worldcup19doha/worldcup2019doha_qual_vt.csv')

fx = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_fx.csv')
fxq = pd.read_csv('../../../scraping/women/worldcup15cottbus/worldcup2015cott_qual_fx.csv')


```

```python
ub.head()
```

```python
ubq.head()
```

```python
ub['round']="final"
ubq['round']="qual"
```

```python
combined_df = pd.concat([ub, ubq], axis=0)
```

```python
ub=combined_df.copy()
```

```python
ub["year"]=2015
```

```python
ub.info()
```

```python
ub.fillna(0, inplace=True)
```

```python
ub['Nation'].unique()
```

```python

# Dictionary mapping nation names to NOC codes
nation_to_noc = {
    'Sweden': 'SWE',
    'Azerbaijan': 'AZE',
    'Portugal': 'POR',
    'Hungary': 'HUN',
    'Spain': 'ESP',
    'Switzerland': 'SUI',
    'Greece': 'GRE',
    'Slovenia': 'SLO',
    'Turkey': 'TUR',
    'Poland': 'POL',
    'Romania': 'ROU',
    'Israel': 'ISR',
    'Germany': 'GER',
    'Norway': 'NOR',
    'Argentina': 'ARG',
    'Czech Republic': 'CZE',
    'Austria': 'AUT',
    'Latvia': 'LAT',
    'Georgia': 'GEO',
    'Finland': 'FIN',
    'Ireland': 'IRL'
}

# Map the "Nations" column to NOC codes
ub['NOC'] = ub['Nation'].map(nation_to_noc)
```

```python
ub=ub.drop(columns=["Nation"])
ub["Nation"] = ub["NOC"]
ub=ub.drop(columns=["NOC"])
```

```python
# Remove rows where "Rank" is not NaN
df = df[df['total_rank'].isna()]

# Reset the index if needed
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
dfg = df.copy()
```

```python

import re

dfg['country'] = None

# Iterate through the "Unnamed: 2" column, extract values within parentheses, and store them in "Vault Rank" column
for index, row in dfg.iterrows():
    value = row['Team']
    if value is not None:
        match = re.search(r'\((.*?)\)', value)
        if match:
            Teams_rank = match.group(1)
            dfg.at[index, 'country'] = Teams_rank



# Remove parentheses and their contents from "Unnamed: 2" column
dfg['Team'] = dfg['Team'].apply(lambda x: re.sub(r'\([^)]*\)', '', str(x)))
```

```python
df = dfg.copy()
```

```python
df.columns
```

```python
column_order = ['Team','country', 'fx', 'fx_rank', 'ph', 'ph_rank', 'sr', 'sr_rank', 'vt',
       'vt_rank', 'pb', 'pb_rank', 'hb', 'hb_rank', 'total', 'total_rank'
       ]

# Create a new DataFrame with the desired column order
df = df[column_order]
```

```python
df.head()
```

```python
df = df.drop(columns=["total", "total_rank"])
```

```python
df.fillna(0, inplace=True)
```

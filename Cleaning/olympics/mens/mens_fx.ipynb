```python
import pandas as pd
import numpy as np
```

```python
df12 = pd.read_csv('../../../scraping/men/olympics/olympic2012men_final/Gymnastics_ 2012_Olympics_Men\'s_artistic_floor.csv')
df16 = pd.read_csv('../../../scraping/men/olympics/olympic2016men/Gymnastics_ 2016_Olympics_Men\'s_artistic_floor.csv')
df20 = pd.read_csv('../../../scraping/men/olympics/mens202olympics_fianl/Gymnastics_ 2020_Olympics_Men\'s_artistic_floor.csv')
```

```python
column_name_mapping = {
    'Name': 'Gymnast',
    'Difficulty': 'D Score',
    'Execution': 'E Score',
    'Penalty': 'Pen.',
}

# Use the rename method to change column names
df16.rename(columns=column_name_mapping, inplace=True)
```

```python
df16
```

```python
df16['Nation'] = df16['Gymnast'].str.extract(r'\((.*?)\)')
df16['Gymnast'] = df16['Gymnast'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
df20['Nation'] = df20['Gymnast'].str.extract(r'\((.*?)\)')
df20['Gymnast'] = df20['Gymnast'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
combined_df = pd.concat([df12, df16, df20], axis=0)

combined_df.reset_index(drop=True, inplace=True)
```

```python
df= combined_df.copy()
```

```python
df = df.fillna(0)
```

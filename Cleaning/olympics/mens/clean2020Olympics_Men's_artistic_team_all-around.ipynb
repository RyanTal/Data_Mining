```python
import re

new_df['Vault Rank'] = None

# Iterate through the "Unnamed: 2" column, extract values within parentheses, and store them in "Vault Rank" column
for index, row in new_df.iterrows():
    value = row['Unnamed: 2']
    if value is not None:
        match = re.search(r'\((.*?)\)', value)
        if match:
            vault_rank = match.group(1)
            new_df.at[index, 'Vault Rank'] = vault_rank

# Print the DataFrame with the new "Vault Rank" column
print(new_df[['Team', 'Unnamed: 2', 'Vault Rank']])
```

```python
import re


# Remove parentheses and their contents from "Unnamed: 2" column
new_df['Unnamed: 2'] = new_df['Unnamed: 2'].apply(lambda x: re.sub(r'\([^)]*\)', '', str(x)))

# Print the DataFrame with the modified "Unnamed: 2" column
print(new_df[['Team', 'Unnamed: 2']])
```

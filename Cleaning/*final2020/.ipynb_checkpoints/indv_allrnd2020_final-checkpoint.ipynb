```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../scraping/scrape2020final_indiv_allrnd.csv')
```

```python
df
```

```python
# Extract data within parentheses and store it in a new column
df["Country"] = df['Gymnast'].str.extract(r'\((.*?)\)')

# Remove data within parentheses from original column
df['Gymnast'] = df['Gymnast'].str.replace(r'\(.*?\)', '')
```

```python
df['Gymnast'] = df['Gymnast'].apply(lambda x: x.split('(')[0].strip())
df
```

```python
# Extract data within parentheses and store it in a new column
df["vault_rank"] = df['vault'].str.extract(r'\((.*?)\)')

df['vault'] = df['vault'].apply(lambda x: x.split('(')[0].strip())
df
```

```python
# Extract data within parentheses and store it in a new column
df["unevenbars_rank"] = df['unevenbars'].str.extract(r'\((.*?)\)')

df['unevenbars'] = df['unevenbars'].apply(lambda x: x.split('(')[0].strip())
df.head()
```

```python
# Extract data within parentheses and store it in a new column
df["bb_rank"] = df['bb'].str.extract(r'\((.*?)\)')

df['bb'] = df['bb'].apply(lambda x: x.split('(')[0].strip())
df.head()
```

```python
# Extract data within parentheses and store it in a new column
df["floor_rank"] = df['floor'].str.extract(r'\((.*?)\)')

df['floor'] = df['floor'].apply(lambda x: x.split('(')[0].strip())
df.head()
```

```python
df["floor_rank"] = df["floor_rank"].apply(lambda x: x.replace('=', ''))
```

```python
df["unevenbars_rank"] = df["unevenbars_rank"].apply(lambda x: x.replace('=', ''))
```

```python
df["bb_rank"] = df["bb_rank"].apply(lambda x: x.replace('=', ''))
```

```python
df["vault_rank"] = df["vault_rank"].apply(lambda x: x.replace('=', ''))
df["vault_rank"]
```

```python
new_order=['Rank', 'Gymnast', 'Country', 'vault', 'vault_rank', 'unevenbars', 'unevenbars_rank',
 'bb', 'bb_rank', 'floor', 'floor_rank','Total']

df = df[new_order]

df
```

```python
df.info()
```

```python
df['vault'] = df['vault'].astype(float)
df['vault_rank'] = df['vault_rank'].astype(float)
df['unevenbars'] = df['unevenbars'].astype(float)
df['unevenbars_rank'] = df['unevenbars_rank'].astype(float)
df['bb'] = df['bb'].astype(float)
df['bb_rank'] = df['bb_rank'].astype(float)
df['floor'] = df['floor'].astype(float)
df['floor_rank'] = df['floor_rank'].astype(float)
```

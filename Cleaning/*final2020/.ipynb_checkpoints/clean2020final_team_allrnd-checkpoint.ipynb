```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../scraping/scrape2020final_team_allrnd.csv')
```

```python
df
```

```python
# Select rows 0 and 5 and create a new DataFrame
new_df = pd.concat([df.iloc[0], df.iloc[5], df.iloc[10], df.iloc[15], df.iloc[20],
                    df.iloc[25], df.iloc[30], df.iloc[35]], axis=1).transpose()

# Drop rows 0 and 5 from original DataFrame
df = df.drop([0, 5, 10, 15, 20, 25, 30, 35])

df = df.reset_index(drop=True)

df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
# Extract data within parentheses and store it in a new column
df["Country"] = df['team'].str.extract(r'\((.*?)\)')
df['team'] = df['team'].apply(lambda x: x.split('(')[0].strip())
df
```

```python
df.info()
```

```python
df['vault'] = df['vault'].astype(float)
df['unevenbars'] = df['unevenbars'].astype(float)
df['bb'] = df['bb'].astype(float)
df['floor'] = df['floor'].astype(float)
```

```python
# function calculate 'Total'
def calculate_total(row):
    return row['vault'] + row['unevenbars'] + row['bb'] + row['floor']

# Apply function to each row and update  'Total' column
df['Total'] = df.apply(lambda row: calculate_total(row), axis=1)

df
```

```python
df["Country"].unique()
```

```python
# ap 'Country' to 'rank'
country_rank_mapping = {
    'ROC': 1,
    'USA': 2,
    'GBR': 3,
    'ITA': 4,
    'JPN': 5,
    'FRA': 6,
    'CHN': 7,
    'BEL': 8
}

# Replace 'Country' values with corresponding ranks
df['rank'] = df['Country'].replace(country_rank_mapping)
df
```

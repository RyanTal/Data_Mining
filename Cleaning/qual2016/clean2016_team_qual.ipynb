```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../scraping/women/qual2016/scrape2016_team_qual.csv')
```

```python
df.tail(50)
```

```python

# Select rows 0 and 5 and create a new DataFrame
new_df = pd.concat([df.iloc[0], df.iloc[6], df.iloc[12], df.iloc[18], df.iloc[24],
                   df.iloc[30], df.iloc[36], df.iloc[42], df.iloc[48],
                   df.iloc[54],df.iloc[60],df.iloc[66]], axis=1).transpose()
new_df

# Drop rows 0 and 5 from the original DataFrame
df = df.drop([0, 6,12,18,24,30,36,42,48,54,60,66])

df = df.reset_index(drop=True)
```

```python
new_df
```

```python
df
```

```python
# Extract data within parentheses and store it in a new column
df["Country"] = df['Team'].str.extract(r'\((.*?)\)')

df['Team'] = df['Team'].apply(lambda x: x.split('(')[0].strip())
df
```

```python
# Split the 'Vault_Scores' column into 'vault1_score' and 'vault2_score'
df[['vault1_score', 'vault2_score']] = df['vault_score'].str.split('/', expand=True)
df
```

```python
df[['vault2_score', 'vault_avg_score']] = df['vault2_score'].str.split('Avg:', expand=True)
df
```

```python
# Remove leading/trailing whitespace
df['vault2_score'] = df['vault2_score'].str.strip()
df['vault_avg_score'] = df['vault_avg_score'].str.strip()
df
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
df
```

```python
df['vault_avg_score'] = df['vault_avg_score'].astype(float)
df.info()
```

```python
df['vault2_score'] = df['vault2_score'].astype(float)
df['vault1_score'] = df['vault1_score'].astype(float)
```

```python
df = df.drop(columns=["vault_score"])
df
```

```python
df.info()
```

```python
df["Name"] = df["Team"]
```

```python
df=df.drop(columns=["Team"])
df
```

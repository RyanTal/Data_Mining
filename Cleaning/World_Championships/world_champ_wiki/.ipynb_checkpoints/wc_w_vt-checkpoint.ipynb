```python
df=combined_df.copy()
```

```python
# Function to remove asterisk (*) from a string
def remove_asterisk(value):
    return value.replace('*', '')

# Apply the remove_asterisk function to the specific cell
row_index_to_update = 7
df.at[row_index_to_update, 'D Score2'] = remove_asterisk(df.at[row_index_to_update, 'D Score2'])
```
